### 04) WAP to find the longest word(s) in a file.

```python
fp = open('new.txt')
s = fp.readline()
l = s.split()
longWord = l[0]
for i in l:
    if len(longWord) < len(i) : 
        longWord = i
print(longWord)
fp.close()
```

### 05) WAP to count the no. of lines, words and characters in a given text file.

```python
fp = open('new.txt')
words = 0
lenChar = 0
data = fp.readlines()
for i in data:
    words+=len(i.split())
    lenChar+=len(i)
print(len(data),words,lenChar)
fp.close()
```

### 10) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
count = 0
num = 2
fp = open('primenumbers.txt','w')
while True:
    for i in range(2,num):
        if num % i == 0:
            break
    else:
        count+=1
        fp.write(str(num)+'\n')
    if count == 100:
        break
    num+=1
fp.close()
```
