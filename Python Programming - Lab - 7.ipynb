### 02) WAP to find Maximum and Minimum K elements in a given tuple.
**Sample Input:**
```
T = (7, 2, 9, 4, 1, 5)
K = 2

```
**Sample Output:**
```
Minimum 2 elements: (1, 2)
Maximum 2 elements: (9, 7)
```

````python
t1 = tuple(map(int,input("Enter Data using space:").split()))

max1 = 0
max2 = 0
min1 = max(t1)
min2 = max(t1)

for i in t1:
    if max2<i:
        max1 = max2
        max2 = i
    if min2>i:
        min1 = min2
        min2 = i
ans1 = (min2,min1)
ans2 = (max2,max1)
print(ans1)
print(ans2)
````

### 03) WAP to find tuples which have all elements divisible by K from a list of tuples.
**Sample Input:**
```
tuple_list = [(2, 4, 6), (3, 6, 9), (4, 8, 12), (5, 10, 15)]
K = 2

```
**Sample Output:**
```
Tuples divisible by 2 : [(2, 4, 6), (4, 8, 12)]
```

````python
s = int(input("Enter Size:"))
k = int(input("Enter K:"))
l1 = list()
for i in range(s):
    t1 = tuple(map(int,input("Enter Data using space:").split()))
    l1.append(t1)
ans = list()
for i in l1:
    for j in i:
        if j%k!=0:
            break
    else:
        ans.append(i)
print(ans)
````

### 12) WAP to merge two tuples of equal length alternately.

### 14) WAP to create (index, value) pairs for prime numbers.

**Sample Input:**
```
T = (4, 7, 9, 11, 15)
```
**Sample Output:**
```
((1, 7), (3, 11))
```

````python
t1 = tuple(map(int,input("Enter Data using space:").split()))
ans = list()

for i in range(len(t1)):
    for j in range(2,t1[i]):
        if t1[i]%j==0:
            break
    else:
        t2 = (i,t1[i])
        ans.append(t2)
print(tuple(ans))
````

### 16) WAP to compute column-wise sum from a tuple of lists.

**Sample Input:**
```
T = ([1, 2, 3], [4, 5, 6], [7, 8, 9])
```

**Sample Output:**
```
12 15 18

```

````python
s = int(input("Size:"))
outer = list()
for i in range(s):
    inner = list(map(int,input("Enter Data using space:").split()))
    outer.append(inner)

ans = [0]*len(outer[0])
for i in range(len(outer)):
    for j in range(len(outer[i])):
        ans[j] = ans[j] + outer[i][j]
print(ans)
````
