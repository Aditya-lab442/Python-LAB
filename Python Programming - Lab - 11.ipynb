### Practical 1: Count Lines, Words and Characters

**Problem Definition:**  
Write a Python program to read text from a file and count the total number of lines, words, and characters. The result should be written to an output file.


```python
def countLWC(data):
    words = 0
    lenChar = 0
    
    for i in data:
        words+=len(i.split())
        lenChar+=len(i)
        fp1 = open('ans.txt','w')
        fp1.write(str(len(data))+" "+str(words)+" "+str(lenChar))
        fp1.close()


fp = open('temp1.txt')
countLWC(fp.readlines())
fp.close()
```


### Practical 2: Search a Word in a File

**Problem Definition:**  
Write a Python program to search a given word in a text file and count its occurrences. Store the result in an output file.


```python
word = input("Enter Word:")

fp = open('temp1.txt')
data = fp.read()
fp1 = open('ans.txt','w')
fp1.write(f"{word} At {data.find(word)} -> {data.count(word)} Times")
fp1.close()
fp.close()
```


### Practical 3: Student Marks and Average

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Store marks in a list, calculate the average using a user-defined function, and write student details along with average marks to an output file.


```python
def calculateAVG(avgList):
    sum1 = 0
    for i in avgList:
        sum1+=int(i)
    return sum1/len(avgList)

fp = open('temp1.txt')
data = fp.readlines()
ansTemp = []
for i in data:
    temp = i.split()
    avgMarks = calculateAVG([temp[2],temp[3],temp[4]])
    
    ansTemp.append(i+""+str(avgMarks)+"\n")
fp1 = open('ans.txt','w')
fp1.writelines(ansTemp)
fp1.close()
fp.close()
```


### Practical 4: Maximum and Minimum using Tuple

**Problem Definition:**  
Write a Python program to read numbers from a file, store them in a tuple, and find the maximum and minimum values. Write the result to an output file.


```python
fp = open('temp1.txt')
data = tuple(fp.read().split())
maxNum = int(data[0])
minNum = int(data[0])
for i in data:
    if maxNum < int(i):
        maxNum = int(i)
    elif minNum > int(i):
        minNum = int(i)
fp1 = open('ans.txt','w')
fp1.write(f"Max = {maxNum} , Min = {minNum}")
fp1.close()
fp.close()
```


### Practical 5: Remove Duplicate Words using Set

**Problem Definition:**  
Write a Python program to read words from a file and remove duplicate words using a set. Store the unique words in an output file.


### Practical 6: Word Frequency using Dictionary

**Problem Definition:**  
Write a Python program to count the frequency of each word in a text file using a dictionary and write the result to an output file.


```python
fp = open('temp1.txt')
temp = fp.read()
data = set(temp.split())
fp1 = open('ans.txt','a')
for i in data:
    fp1.write(f"{i}:{temp.count(i)} \n")
fp1.close()
fp.close()
```


### Practical 7: Student Result and Grade Calculation

**Problem Definition:**  
Write a Python program to read roll number, name, and marks of students from a file. Use a user-defined function to determine grade (A/B/C/Fail) and write the student result to an output file.

**Grade Criteria:**  
- Marks ≥ 75 : Grade A  
- Marks ≥ 60 : Grade B  
- Marks ≥ 40 : Grade C  
- Marks < 40 : Fail


```python
def calculateGrade(Marks):
    if int(Marks) >= 75:
        return "Grade A"
    elif int(Marks) >= 60:
        return "Grade B"
    elif int(Marks) >= 40:
        return "Grade C"
    else:
        return "Fail"
fp = open('temp1.txt')
fp1 = open('ans.txt','a')
data = fp.readlines()
for i in data:
    grade = calculateGrade(i.split()[2])
    fp1.write(grade+'\n')
fp1.close()
fp.close()
```
