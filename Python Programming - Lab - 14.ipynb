### 04) WAP to plot a line chart showing temperature variation during a week. Mark the lowest and highest temperatue.

```python
Days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
Temperature = [20, 32, 31, 33, 34, 43, 36]
```

```python
plt.plot(Days,Temperature,marker="o")
plt.annotate("MAX Temp",xy=["Fri",max(Temperature)])
plt.annotate("  MIN",xy=["Mon",min(Temperature)])
plt.show()
```

### 05) WAP to create a detailed pie chart representing the monthly expenditure distribution of a family. 
### The chart should display labels, percentage values, title, different colors for each slice, and explode one slice to highlight the highest expense.

```python
Categories = ["Rent", "Food", "Education", "Transport", "Entertainment", "Savings"]
Amount = [15000, 8000, 6000, 3000, 2000, 5000]
```

```python
plt.pie(Amount,labels = Categories,autopct="%.2f%%",explode=[0.1,0,0,0,0,0])
plt.title("PIE CHART")
plt.show()
```

### 06) WAP to create bar chart showing the marks obtained by students in different subjects.
