### 10) WAP to accept a float number and display with 2 decimals, 3 decimals, and width 10.
#### Input : 37.2567
#### Output :
##### 2 decimals: 37.26
##### 3 decimals: 37.257
##### Width 10:      37.26

```python
a = input("Number:")
a = float(a)
print(f"2 Decimal:{a:.2f}")
print(f"3 Decimal:{a:.3f}")
print(f"Width:{a:10.2f}")
```
