### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
def calculateBMI(m,h):
    return m/h**2


mass = int(input("Enter mass:"))
height = int(input("Enter Height:"))

BMI = calculateBMI(mass,height)

print("BMI = ",BMI)
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def isPrime(n):
    for i in range(2,n):
        if n%i == 0:
            return 0
    else:
        return 1

num = int(input("Enter Number:"))

if isPrime(num):
    print("Prime Number")
else:
    print("Not Prime Number")
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
num1 = int(input("Enter Number:"))
num2 = int(input("Enter Number:"))

def isPrime(n):
    for i in range(2,n):
        if n%i == 0:
            return 0
    else:
        return n

def rangePrime(num1,num2):
    l1 = []
    for i in range(num1,num2+1):
        if isPrime(i) != 0:
            l1.append(isPrime(i))
    return l1
print(rangePrime(num1,num2))
```

### 05) Write a function that returns True if the given string is Palindrome or False otherwise.

```python
def palindrome(s):
    if s == s[-1::-1]:
        return True
    else:
        return False

s = input("Enter String:")

if palindrome(s):
    print("Palindrome")
else:
    print("Not Palindrome")
```

### 07) Write a function to calculate the sum of the first element of each tuples inside the list.

```python
def sumFirstEle(l1):
    sum1 = 0
    for i in l1:
        sum1+=i[0]
    return sum1

n = int(input("Enter size of list"))
l1 = []
for i in range(n):
    temp = tuple(map(int,input("Enter Element With Space").split(" ")))
    l1.append(temp)
print(sumFirstEle(l1))
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python

def fibo(n):
   if n <= 1:
       return n
   else:
       return fibo(n-1) + fibo(n-2)
       
n = int(input("Enter N:"))


print("Fibonacci sequence:")
for i in range(n):
    print(fibo(i),end = " ")
```

### 09) Write a function to get the name of the student based on the given rollno.
#### Example: Given dict1 = {101:'Ajay', 102:'Rahul', 103:'Jay', 104:'Pooja'} find name of student whose rollno = 103

```python
def findName(d,n):
    return d[n]

size = int(input("Enter size:"))

d1 = {}

for i in range(size):
    j,k = input("Enter key : value : ").split(" ")
    d1[int(j)] = k

num = int(input("Enter Roll No. for find:"))

name = findName(d1,num)

print(name)
```

### 10) Write a function to get the sum of the scores ending with zero.
#### Example : scores = [200, 456, 300, 100, 234, 678] 
#### Ans = 200 + 300 + 100 = 600

```python
def sumEndZero(l1):
    total = 0
    for i in l1:
        if i % 10 == 0:
            total +=i
    return total

l1 = list(map(int,input("Enter Element With Space").split(" ")))
print("Ans = ",sumEndZero(l1))
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def invert(d):
    ans = {}
    for i,j in d.items():
        ans[j] = i
    return ans

size = int(input("Enter size:"))

d1 = {}

for i in range(size):
    j,k = input("Enter key : value : ").split(" ")
    d1[j] = int(k)

ans = invert(d1)
print("Before = ",d1)
print("After =",ans)
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def findUpperAndLower(s):
    ucount = 0
    lcount = 0
    for i in s:
        if i.isupper():
            ucount+=1
        else:
            lcount+=1
    return ucount,lcount
s = input("Enter String:")

ans = findUpperAndLower(s)

print("no_upper = ",ans[0],"no_lower = ",ans[1])
```

### 13) Write a lambda function to get smallest number from the given two numbers.

```python
num1,num2 = input("Enter Two Number").split(" ")

findMin = lambda n1,n2: n1 if n1<n2 else n2

print(findMin(int(num1),int(num2)))
```

### 14) For the given list of names of students, extract the names having more that 7 characters. Use filter().

```python
def findName(s):
    if len(s)>7:
        return s

l1 = input("Enter Name:").split()

ans = list(filter(findName,l1))

print(ans)
```

### 15) For the given list of names of students, convert the first letter of all the names into uppercase. use map().

```python
def convert(s):
    return s.title()

l1 = input("Enter Name:").split()

ans = list(map(convert,l1))

print(ans)
```

### 16) Write udfs to call the functions with following types of arguments:
1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable Legngth Positional(*args) & variable length Keyword Arguments (**kwargs)
5. Keyword-Only & Positional Only Arguments

```python
def pos(a,b):
    return a + b
def keyWord(a,b):
    return a + b
def defau(a = 0,b = 0):
    return a + b
def vrps(*args):
    total = 0
    for i in args:
        total += i
    return total
def vrkw(**kwargs):
    ans = {}
    for i,j in kwargs.items():
        # print(f"{i} = {j}")
        ans[i] = j
    return ans
def pson(a,b,/):
    return a + b
def kwon(*,a,b):
    return a + b
num1,num2 = input("Enter Two Number").split(" ")

print("Positional = ",pos(int(num1),int(num2)))
print("Keyword = ",keyWord(b = int(num2),a = int(num1)))
print("default = ",defau(int(num1)))
print("Variable Positional = ",vrps(int(num1),int(num2),4,5,6))
print("Variable Keyword = ",vrkw(a = 1,b=2,c=3))
print("Positional only = ",pson(int(num1),int(num2)))
print("Keyword Only = ",kwon(b = int(num2),a = int(num1)))
```
