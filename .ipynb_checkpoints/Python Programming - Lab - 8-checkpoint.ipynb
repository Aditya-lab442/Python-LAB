### 01) WAP to find their union, intersection, and difference of the given two sets.
**Sample Input:**
```
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

```
**Sample Output:**
```
Union: {1, 2, 3, 4, 5, 6}
Intersection: {3, 4}
Difference (A-B): {1, 2}

```

````python
a=set(map(int,input("enter data:").split()))
b=set(map(int,input("enter data:").split()))

print(f"Union: {a.union(b)} \n intersection: {a.intersection(b)} \n Difference: {a.difference(b)}")
````

### 03) WAP to remove duplicate elements from a list while preserving unique values.
**Sample Input:**
```
[10, 20, 10, 30, 20, 40]

```
**Sample Output:**
```
{10, 20, 30, 40}
```

````python
a=list(map(int,input("enter data:").split()))
b=set()

for i in a:
    if i not in b:
        b.add(i)
print(b)
````

### 05) WAP to Merge two dictionaries. If a key exists in both dictionaries, add their values.

**Sample Input:**
```
Dict1 = {1: 10, 2: 20, 3: 30}
Dict2 = {2: 40, 3: 50, 4: 60}

```
**Sample Output:**
```
{1: 10, 2: 60, 3: 80, 4: 60}
```

````python
data=input("Enter dict1 key-value pairs: ").split()
dict1={int(data[i]): int(data[i+1]) for i in range(0, len(data), 2)}
data=input("Enter dict2 key-value pairs: ").split()
dict2={int(data[i]): int(data[i+1]) for i in range(0, len(data), 2)}
dict3=dict1.copy()

for k1,v1 in dict2.items():
    if k1 in dict3:
        dict3[k1]+=v1
    else:
        dict3[k1]=v1
print(dict3)
````

### 06) WAP to Sort a dictionary in ascending order of its values.
Hint: in sorted(), use udf for key parameter


**Sample Input:**
```
{'Math': 70, 'Physics': 85, 'Chemistry': 60}

```
**Sample Output:**
```
{'Chemistry': 60, 'Math': 70, 'Physics': 85}
```

````python
data=input("Enter dict1 key-value pairs: ").split()
dict1={data[i]: int(data[i+1]) for i in range(0, len(data), 2)}
items=list(dict1.items())
for i in range(len(items)):
    for j in range(i+1,len(items)):
        if items[i][1]>items[j][1]:
            items[i],items[j]=items[j],items[i]
dict2=dict(items)
print(dict2)
````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```
**Sample Output:**
```
Max Value Key: B
Min Value Key: C

```

````python
data=input("Enter dict1 key-value pairs: ").split()
dict1={data[i]: int(data[i+1]) for i in range(0, len(data), 2)}
mn=min(dict1.values())
mx=max(dict1.values())
for k,v in dict1.items():
    if v==mn:
        print(f"min value key: {k}")
    elif v==mx:
        print(f"max value key: {k}")
````

### 11) WAP to find the student with the highest total marks, given student names and their marks.

**Sample Input:**
```
{'Amit': [70, 80, 90], 'Riya': [85, 75, 80], 'Neha': [60, 70, 65]}
```
**Sample Output:**
```
Topper: Amit
```

````python
data=input("Enter student names followed by their marks: ").split()
dict1={}
i=0
while i<len(data):
    name=data[i]
    marks=[]
    i+=1
    while i<len(data) and data[i].isdigit():
        marks.append(int(data[i]))
        i+=1
    dict1[name]=marks
topper=''
highest_total=0

for name, marks in dict1.items():
    total=sum(marks)
    if total>highest_total:
        highest_total=total
        topper=name
print("Topper:", topper)
````

### 12) WAP to extract keys whose values are even numbers.

**Sample Input:**
```
{1: 11, 2: 20, 3: 15, 4: 40}
```
**Sample Output:**
```
[2, 4]
```

````python
data=input("Enter dict1 key-value pairs: ").split()
dict1={int(data[i]): int(data[i+1]) for i in range(0, len(data), 2)}
even_keys=[k for k, v in dict1.items() if v % 2 == 0]
print(even_keys)
````

```
dict1 = {'a': 5, 'c': 8, 'e': 2}
```

### 15) WAP to create a dictionary using dictionary comprehension from two lists containing student names and marks. Include only those students in the dictionary whose marks are greater than or equal to 75.

**Sample Input:**
```
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]
```
**Sample Output:**
```
{'Amit': 78, 'Neha': 85}
```

````python
students = ['Amit', 'Riya', 'Neha', 'Karan']
marks = [78, 72, 85, 60]
ans = {students[i]:marks[i] for i in range(len(students)) if marks[i]>=75}
print(ans)
````
