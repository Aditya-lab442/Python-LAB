### 1) Character Category Counter
**Problem:** Count uppercase, lowercase, digits, and special characters in a string.

**Sample Inputs & Outputs:**
- Input: `Python3@Lab` → Uppercase: 2, Lowercase: 7, Digits: 1, Special: 1
- Input: `HELLO123` → Uppercase: 5, Lowercase: 0, Digits: 3, Special: 0
- Input: `hi@#9` → Uppercase: 0, Lowercase: 2, Digits: 1, Special: 2

```python
s = input("Enter String");
uc = 0;
lc = 0;
d = 0;
sc = 0;
for i in s:
    if i.islower():
        lc = lc+1;
    elif i.isupper():
        uc = uc+1;
    elif i.isdigit():
        d = d+1;
    else:
        sc = sc + 1;
        
print("Uppercase:",uc,"Lowercase:",lc,"Digit:",d,"Special:",sc);
```

### 2) Reverse Each Word
- Input: `Python is Easy` → `nohtyP si ysaE`
- Input: `Hello World` → `olleH dlroW`
- Input: `Learn Python Fast` → `nraeL nohtyP tsaF`

```python
s = input("Enter String:")
rs = s.split(" ")
ans = ""
for i in rs:
    ans = ans+(i)[::-1]
    ans = ans + " "
print(ans)
```

### 8) Replace Vowels
- `Programming` → `Pr*gr*mm*ng`
- `Education` → `*d*c*t**n`
- `Sky` → `Sky`

```python
s = input("Enter String:")

for i in s:
    if i.lower()=='a' or i.lower()=='e' or i.lower()=='i' or i.lower()=='o' or i.lower()=='u':
        s = s.replace(i,'*')
print(s)
```

### 9) String Compression
- `aaabbcccc` → `a3b2c4`
- `xxxyyyz` → `x3y3z1`
- `abcd` → `a1b1c1d1`

```python
s = input("Enter String:")
ans = ""
count = 1;
for i in range(1,len(s)):
    if s[i-1] == s[i]:
        count = count+1;
    else:
        ans = ans+ s[i-1]+str(count)
        count = 1;
        

ans = ans+ s[-1]+str(count)
print(ans)
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
s = input("Enter String:")
ans = ""
for i in s:
    if i.islower():
        ans = ans + i.upper()
    else:
        ans = ans + i.lower()
print(ans)
```

### 13) Longest Word Finder
- `Python programming is interesting` → programming
- `I love coding` → coding
- `Data structures and algorithms` → structures

```python
s = input("Enter String:")
max1 = 0;
l1 = s.split(" ")

ans = ""

for i in l1:
    if max(max1,len(i)) > max1:
        ans = i
        max1 = max(max1,len(i))
print(ans)
```
